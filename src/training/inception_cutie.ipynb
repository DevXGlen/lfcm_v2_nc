```python
import inceptionv3Cutie
import torch
import torch.nn.functional as F
```

```python
# inceptionv3 BasicConv2d
inceptionv3Cutie.BasicConv2d(3, 32, kernel_size=3, stride=2)
```

```python
# testing BasicConv2d 
# depth, height, width
# 3 x 299 x 299
tensor = torch.rand(3, 299, 299)
conv2d = inceptionv3Cutie.BasicConv2d(3, 32, kernel_size=3, stride=2)
conv2d = conv2d(tensor.clone().unsqueeze(0))
print(conv2d)
```

```python
# testing flatten

tensor = torch.rand(3 , 5, 6)

tensor_view = tensor.detach().clone().view(tensor.size(0), -1)
tensor_flatten = torch.flatten(tensor.detach().clone(), 1)

#print('tensor:', tensor)
#print('tensor_view:', tensor_view)
print('tensor view size:', tensor_view.size())
print('tensor flatten size:', tensor_flatten.size())
#print('tensor_flatten:', tensor_flatten)
print(tensor_view == tensor_flatten)
```

```python
# InceptionAux 
# Default tensor dimension used in InceptionAux
# batch x depth x height x width
tensor = torch.rand(2, 768, 299, 299)
# inceptionv3Cutie.InceptionAux(in_channels, num_channels)
inceptionAux = inceptionv3Cutie.InceptionAux(768, 1000)
inceptionAux = inceptionAux(tensor)

print(inceptionAux)
```

```python
# InceptionE 
# Default tensor dimension used in InceptionE
# batch x depth x height x width
tensor = torch.rand(2, 2048, 299, 299)
# inceptionv3Cutie.InceptionE(in_channels)
inceptionE = inceptionv3Cutie.InceptionE(2048)
inceptionE = inceptionE(tensor.detach().clone())

print(inceptionE)
```

```python
# InceptionD 
# Default tensor dimension used in InceptionD
# batch x depth x height x width
tensor = torch.rand(2, 768, 299, 299)
# inceptionv3Cutie.InceptionD(in_channels)
inceptionD = inceptionv3Cutie.InceptionD(768)
inceptionD = inceptionD(tensor.detach().clone())

print(inceptionD)
```

```python
# InceptionC 
# Default tensor dimension used in InceptionC
# batch x depth x height x width
tensor = torch.rand(2, 768, 299, 299)
# inceptionv3Cutie.InceptionC(in_channels)
inceptionC = inceptionv3Cutie.InceptionC(768, channels_7x7=128)
inceptionC = inceptionC(tensor.detach().clone())

print(inceptionC)
```

```python
# InceptionB 
# Default tensor dimension used in InceptionB
# batch x depth x height x width
tensor = torch.rand(4, 288, 299, 299)
# inceptionv3Cutie.InceptionB(in_channels)
inceptionB = inceptionv3Cutie.InceptionB(288)
inceptionB = inceptionB(tensor.detach().clone())

print(inceptionB)
```

```python
# InceptionA 
# Default tensor dimension used in InceptionA
# batch x depth x height x width
tensor = torch.rand(4, 192, 299, 299)
# inceptionv3Cutie.InceptionA(in_channels)
inceptionA = inceptionv3Cutie.InceptionA(192, pool_features=32)
inceptionA = inceptionA(tensor.detach().clone())

print(inceptionA)
```

```python
# Inception3 
# Default tensor dimension used in Inception3
# batch x depth x height x width
tensor = torch.rand(2, 3, 299, 299)
# inceptionv3Cutie.Inception3()
inception3 = inceptionv3Cutie.Inception3(num_classes=2)
inception3 = inception3(tensor.detach().clone())

print(inception3)
print(inception3[0].size())
```

```python
# Inception3 (Pretrained)
# Default tensor dimension used in Inception3
# batch x depth x height x width
tensor = torch.rand(2, 3, 299, 299)
# inceptionv3Cutie.Inception3()
pretrainedInception3 = inceptionv3Cutie.customInception3(pretrained=True, aux_logits=False)
pretrainedInception3 = pretrainedInception3(tensor.detach().clone())

print(pretrainedInception3)
print(pretrainedInception3.size())
```
